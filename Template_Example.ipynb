```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('dbz.jpg')
cv2_imshow(image)
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) 
```

```python
#Load our template image
dbzcrop = cv2.imread('dbzcrop.jpg',0)
#cv2_imshow(dbzcrop)
result = cv2.matchTemplate(gray,dbzcrop,cv2.TM_CCOEFF)
min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(result)

```

```python
#CREATING BOUNDING BOX
top_left=max_loc
bottom_right=(top_left[0]+100,top_left[1]+100) 
#we can increase values of +50,+100,etc etc to adjust the area of bounding box needed
cv2.rectangle(image,top_left,bottom_right,(0,255,0),5)

cv2_imshow(image)
```
